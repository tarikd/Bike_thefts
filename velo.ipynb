### American cities like Seattle et Philadelphia have open data portals providing access to hundred of data sets. I ran into Philadelphia's bicycle thefts dataset and I decided to see if there is a correlation between the climate conditions and the spike in bike thefts.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

#### Let's take a look at the CSV

```python
df = pd.read_csv('bikes.csv')
df.head()
```

#### The columns that I want to use are the location where the bicycles have been stolen, the theft date, year and hour, the value of the stolen bicycle and the value code.

```python
df = df[['LOCATION_B', 'THEFT_DATE', 'THEFT_YEAR', 'THEFT_HOUR', 'STOLEN_VAL', 'UCR']]
df.columns = ['Location', 'Date', 'Year', 'Hour', 'Value', 'Value_Code']
df.head()
```

#### I will now create a plot of the thefts over the entire time period.

```python
pd.options.display.mpl_style = 'default'

rcParams['figure.figsize'] = (20, 8)
rcParams['figure.dpi'] = 300
```

```python
df['Date'] = pd.to_datetime(pd.Series(df['Date']))
df.set_index('Date', drop=False, inplace=True)

df['Thefts'] = df['Date'].value_counts()

plt.figure()
plt.ylim(0,20)

df['Thefts'].plot(x_compat=True)
```

#### Something that stood out immediately was how many more thefts occured in the warmer months than in the winter. Which seems coherent with the fact that the summer is more proper for a bike ride.

#### After seeing this coorelation I downloaded some weather data for Philadelphia going back to January 1st, 2010(where the theft reporting begins) and loaded it in.

```python
temp = pd.read_csv('climate.csv',
            parse_dates={'Date': ['MO','DAY','YEAR']},
            usecols=['MO', 'DAY', 'YEAR', 'HIGH'])

temp.head()
```

```python
temp.plot()
```

#### This temperature plot seems similar to the theft plot.

#### We merge the thefts and the climate tables using the Date column.

```python
merged = pd.merge(df, temp, on='Date', how='outer')
merged.columns = ['Location','Date','Year','Hour','Value','Value_Code','Thefts','Temp']

merged.head()
```

```python
new = merged.sort('Thefts', ascending=False)
# Return new without null values
new = new.dropna()
new.head()
```
